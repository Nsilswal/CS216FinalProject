```python
# Importing packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy as sci
```

```python
# Importing Data
education = pd.read_csv("education.csv")
vaccine = pd.read_csv("vaccinesbycounty.csv")
income = pd.read_csv("MedianIncomeByCounty.csv")
demVsrep = pd.read_csv("democratic_vs_republican_votes_by_usa_state_2020.csv")
```

```python
# ViewData
education
```

```python
education = education.iloc[3:]
ed = education.reset_index()
ed.drop(columns = ['index'])
ed.rename(columns=ed.iloc[0], inplace = True)
ed = ed.drop(3, axis = 1)
ed = ed.dropna()
ed = ed.drop(0)
ed =ed.reset_index(drop =True)
ed = ed.rename(columns={'Federal Information Processing Standards (FIPS) Code': 'FIPS'})
ed['FIPS'] = ed['FIPS'].astype(np.int64)
ed
```

```python
vaccine
```

```python
vaccine = vaccine[vaccine["FIPS"].str.contains("unk")==False]
vaccine = vaccine[vaccine["FIPS"].str.contains("UNK")==False]
vaccine = vaccine[vaccine['Recip_State'].str.contains('PR') == False]
vaccine = vaccine[vaccine['Series_Complete_Yes'] > 0]
vaccine = vaccine[vaccine['Series_Complete_Pop_Pct'] > 0]
vaccine['Pop'] = round(vaccine['Series_Complete_Yes']*100/vaccine['Series_Complete_Pop_Pct'])
vaccine['FIPS'] = vaccine['FIPS'].astype(np.int64)
vaccine['Series_Complete_Pop_Pct'] = vaccine['Series_Complete_Pop_Pct'].astype(np.double)
vaccine = vaccine.reset_index(drop = True)
vaccine
#need to create new table where vaccine percentage by state b/c voting is by state use GroupBy
```

```python
demVsrep['Total'] = demVsrep['DEM'] + demVsrep["REP"]
demVsrep['TotalDem'] = demVsrep['percent_democrat']* demVsrep["Total"]/100
demVsrep
```

```python
income = income[['FIPS_code','State','Area_name','Median_Household_Income_2020']]
income = income[income['Area_name'].str.contains(",") == True]
income = income[income['State'] != "PR"]
income = income.dropna()
income['Median_Household_Income_2020'] = income['Median_Household_Income_2020'].str.replace(',','')
income['Median_Household_Income_2020'] = income['Median_Household_Income_2020'].astype(np.int64)
income = income.reset_index(drop =True)
income = income.rename(columns={'FIPS_code': 'FIPS'})
income
```

```python
vaccineMean = vaccine['Series_Complete_Yes'].sum()/(vaccine['Pop'].sum())
demMean = demVsrep['TotalDem'].sum()/demVsrep['Total'].sum()
print("VaccineMean%: ",vaccineMean)
print('DemMean% :', demMean)
```

```python
merge = pd.merge(vaccine, income, on = 'FIPS', how = 'inner')
merge
```

```python
plt.scatter(merge['Median_Household_Income_2020'],merge['Series_Complete_Pop_Pct'])
m, b = np.polyfit(merge['Median_Household_Income_2020'], merge['Series_Complete_Pop_Pct'], 1)
plt.plot(merge['Median_Household_Income_2020'], m*merge['Median_Household_Income_2020']+b)
plt.xlabel("Median Household Income 2020 ($)")
plt.ylabel("Percent Vaccinated")
plt.title("Percent Vaccinated VS Median Household Income")
```

```python
slope, intercept, r_value, p_value, std_err = sci.stats.linregress(merge['Median_Household_Income_2020'], merge['Series_Complete_Pop_Pct'] )
rsquared = r_value**2
print("the slope is " , slope) 
print("the intercept is " , intercept)
print("the r_value is " , r_value)
print("the rsquared is " , rsquared)
print("the p_value is " , p_value)
print("the std_err is " , std_err)
```

```python
mergeEd = pd.merge(vaccine, ed, on = 'FIPS', how = 'inner')
mergeEd = mergeEd.drop('Unnamed: 5', axis = 1)
mergeEd['Percent of adults with a high school diploma only, 2016-20'] = mergeEd['Percent of adults with a high school diploma only, 2016-20'].astype(np.double)
mergeEd['Percent of adults with less than a high school diploma, 2016-20'] = mergeEd['Percent of adults with less than a high school diploma, 2016-20'].astype(np.double)
mergeEd['% HS Diploma or below'] = mergeEd['Percent of adults with less than a high school diploma, 2016-20']+mergeEd['Percent of adults with a high school diploma only, 2016-20']
```

```python
plt.scatter(mergeEd['% HS Diploma or below'],mergeEd['Series_Complete_Pop_Pct'])
m, b = np.polyfit(mergeEd['% HS Diploma or below'], mergeEd['Series_Complete_Pop_Pct'], 1)
plt.plot(mergeEd['% HS Diploma or below'], m*mergeEd['% HS Diploma or below']+b)
plt.xlabel('% HS Diploma or below')
plt.ylabel("% Vaccinated")
plt.title("% Vaccinated VS % HS Diploma or below")
```

```python
slope, intercept, r_value, p_value, std_err = sci.stats.linregress(mergeEd['Percent of adults with a high school diploma only, 2016-20'], mergeEd['Series_Complete_Pop_Pct'] )
rsquared = r_value**2
print("the slope is " , slope) 
print("the intercept is " , intercept)
print("the r_value is " , r_value)
print("the rsquared is " , rsquared)
print("the p_value is " , p_value)
print("the std_err is " , std_err)
```

```python
mergeEd["Percent of adults completing some college or associate's degree, 2016-20"] = mergeEd["Percent of adults completing some college or associate's degree, 2016-20"].astype(np.double)
mergeEd["Percent of adults with a bachelor's degree or higher 2016-20"] = mergeEd["Percent of adults with a bachelor's degree or higher 2016-20"].astype(np.double)
mergeEd['% Some College or above'] = mergeEd["Percent of adults completing some college or associate's degree, 2016-20"]+mergeEd["Percent of adults with a bachelor's degree or higher 2016-20"]
```

```python
slope, intercept, r_value, p_value, std_err = sci.stats.linregress(mergeEd['% Some College or above'], mergeEd['Series_Complete_Pop_Pct'] )
rsquared = r_value**2
print("the slope is " , slope) 
print("the intercept is " , intercept)
print("the r_value is " , r_value)
print("the rsquared is " , rsquared)
print("the p_value is " , p_value)
print("the std_err is " , std_err)

```

```python
vaccineGroup = vaccine.copy()
vaccineGroup['Series_Complete_Pop'] = vaccineGroup['Pop']*vaccineGroup['Series_Complete_Pop_Pct']/100
vaccineGroup = vaccineGroup.drop(['Recip_County','Unnamed: 5', 'FIPS', 'Series_Complete_Pop_Pct', 'Series_Complete_Yes'], axis = 1)
vaccineGroup= vaccineGroup.groupby(['Recip_State']).sum()
vaccineGroup['%vaccinated'] = vaccineGroup['Series_Complete_Pop']/vaccineGroup['Pop']*100
vaccineGroup = vaccineGroup.reset_index()
vaccineGroup = vaccineGroup.rename(columns={'Recip_State': 'usa_state_code'})

vaccineGroup
```

```python
mergeVote = pd.merge(vaccineGroup, demVsrep, on = 'usa_state_code', how = 'inner')
mergeVote
```

```python
plt.scatter(mergeVote['percent_democrat'],mergeVote['%vaccinated'])
m, b = np.polyfit(mergeVote['percent_democrat'], mergeVote['%vaccinated'], 1)
plt.plot(mergeVote['percent_democrat'], m*mergeVote['percent_democrat']+b)
plt.xlabel('% Democrat')
plt.ylabel("% Vaccinated")
plt.title("% Vaccinated VS % Democrat")
```

```python
slope, intercept, r_value, p_value, std_err = sci.stats.linregress(mergeVote['percent_democrat'], mergeVote['%vaccinated'] )
rsquared = r_value**2
print("the slope is " , slope) 
print("the intercept is " , intercept)
print("the r_value is " , r_value)
print("the rsquared is " , rsquared)
print("the p_value is " , p_value)
print("the std_err is " , std_err)
```
